```python
import pandas as pd

# Load the dataset
file_path = "/content/fifa_players.csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
df.info(), df.head()
```

```python
# Step 1: Handle missing values

# Check missing values per column
missing_values = df.isnull().sum()

# Drop columns with excessive missing values (e.g., national team-related columns)
drop_columns = ["national_team", "national_rating", "national_team_position", "national_jersey_number"]
df_cleaned = df.drop(columns=drop_columns)

# Fill missing financial values (value_euro, wage_euro, release_clause_euro) with median values
financial_cols = ["value_euro", "wage_euro", "release_clause_euro"]
for col in financial_cols:
    df_cleaned[col] = df_cleaned[col].fillna(df_cleaned[col].median())

# Verify missing values after handling
df_cleaned.isnull().sum()
```
