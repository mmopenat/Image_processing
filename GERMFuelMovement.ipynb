```python
import pandas as pd

# Load the data from the uploaded file
file_path = '/content/GermFuelMovement.csv'
fuel_data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
fuel_data.head()
```

```python
fuel_data = pd.read_csv(file_path, delimiter=';')

# Display the first few rows to confirm
fuel_data.head()
```

```python
# Now we will calculate the total FUELMOVELITERS per SITENAME, per REGION, per day.
# First, let's convert the `EVENTTIME` to datetime format and extract the date.

fuel_data['EVENTTIME'] = pd.to_datetime(fuel_data['EVENTTIME'])
fuel_data['EVENT_DATE'] = fuel_data['EVENTTIME'].dt.date

# Grouping the data by 'SITENAME', 'REGION', and 'EVENT_DATE', and summing up 'FUELMOVELITERS'
fuel_summary = fuel_data.groupby(['SITENAME', 'REGION', 'EVENT_DATE'])['FUELMOVELITERS'].sum().reset_index()

# simply use print in a script
print(fuel_summary)
```

```python

# Save the file to the current directory
output_file_path = 'Fuel_Movement_Summary.xlsx'
fuel_summary.to_excel(output_file_path, index=False)

# Confirm the save path
output_file_path

```
